## Описание файла AppleStore.csv:

    - "n": номер приложения 
    - "id" : идентификатор приложения
    - "track_name": название приложения
    - "size_bytes": размер в байтах
    - "currency": тип валюты, в которой измеряется цена
    - "price": цена
    - "rating_count_tot": суммарное число оценок пользователей (для всех версий)
    - "rating_count_ver": суммарное число оценок пользователей (для текущей версии)
    - "user_rating" : средняя оценка пользователей (для всех версий)
    - "user_rating_ver": средняя оценка пользователей (для текущей версии)
    - "ver": код последней версии 
    - "cont_rating": возрастное ограничение контента
    - "prime_genre": основной жанр 
    - "sup_devices.num": количество поддерживающих устройств
    - "ipadSc_urls.num": количество скриншотов в описании
    - "lang.num": число поддерживаемых языков
    - "vpp_lic": лицензирование на основе устройства Vpp

```python
# Переименуем столбец "user_rating" в "userrating"
df = df.rename(columns={'user_rating': 'userrating'})

# Переименуем столбцы так, чтобы они все начинались с заглавной буквы
column_dict = {}
for i in df.columns:
    column_dict[i] = i.capitalize()
df = df.rename(columns=column_dict)
df
```

```python
# Отобразим строки, которые соответствуют приложениям со средней оценкой пользователей (для всех версий) выше 4
df[df['Userrating'] > 4]
```

```python
# Отобразим строки, которые соответствуют приложениям со средней оценкой пользователей (для текущей версии) равной 4
df[df['User_rating_ver'] == 4]
```

```python
# Отобразим строки, которые соответствуют приложениям с суммарным числом оценок
# пользователей (для всех версий) от 100000 до 200000 (включительно)
df[(100000 <= df["Rating_count_tot"]) & (df["Rating_count_tot"] <= 200000)]
```

```python
# Отобразим строки, которые соответствуют приложениям с возрастным ограничением 
# контента 12+ или 17+ и ценой выше 5 долларов

df.query("Cont_rating in ('12+', '17+')")[(df['Currency'] == 'USD') & (df['Price'] > 5)]
```

```python
# Добавим в таблицу новый столбец, который будет представлять собой общую среднюю 
# оценку приложения, рассчитанную как среднее арифметическое от средней оценки
# пользователей для всех версий и для текущей версии

df["UserRatingAvg"] = ((df["Userrating"] + df["User_rating_ver"]) / 2).round(2)
df
```

```python
# Рассчитаем среднюю цену приложений разного жанра

df.groupby("Prime_genre")[["Price"]].mean()
```

```python
# Рассчитаем среднюю оценку пользователей (для всех версий)
# по жанрам отдельно для платных и бесплатных приложений

df['IsFree'] = df.Price == 0
average_price_by_genre = df.groupby(['Prime_genre', 'IsFree'])['User_rating_ver'].min()
average_price_by_genre
```

```python
# Рассчитаем медиану цены приложений, у которых число поддерживаемых языков превышает 2

df[df["Lang.num"] > 2]["Price"].median()
```

```python
# Создадим новый столбец, в которому будет указана цена приложений в рублях 
# по сегодняшнему обменному курсу валют

def getUsdRate():
    import requests
    import xml.etree.ElementTree as ET

    data = requests.get('http://www.cbr.ru/scripts/XML_daily.asp?numcode=051').content
    tree = ET.ElementTree(ET.fromstring(data))
    rate = float(tree.getroot().findall('.//Valute[CharCode="USD"]/VunitRate')[0].text.replace(',','.'))
    return rate

rate = getUsdRate()
df["PriceRub"] = df["Price"] * getUsdRate()
df
```

```python
# Построим диаграмму, демонстрирующую взаимосвязь суммарного числа оценок пользователей
# (для текущей версии) и средней оценки пользователей (для текущей версии) 

plot = df.plot.scatter(x = 'Rating_count_ver', y = 'User_rating_ver', title = "Взаимосвязь рейтинг и количества оценок" )
plot.set_xlabel("Рейтинг для текущей версии")
plot.set_ylabel("Число оценок текущей версии")


```

- Точечная диаграмма позволяет увидеть коррелируют между собой данные или нет

- Однако, из низкого числа оценок не следует низкая средняя оценка
